## 2.1 Claude Code 工作流

```python
Claude_Code_Flow = '''
1. 启动 (Launch): claude 命令启动，自动检测 .claude .md 配置
   /cliprompt 加载操作指导能力

2. 议记录 (Session): .claude/history/ 持久化存储
   CREATE/READ/WRITE 议记录来回顾代码内容

3. 代码生成 (Code Gen):
   PRINT 模式: 生成内容打印到标准输出，用户确认后再写入
   AUTO-EDIT 模式: 直接修改文件（需明确 --approve-edits）

4. Git 集成: 自动 commit，提示涉变曲线
   支持 GitHub PR 并入

5. 环执行 (Sandbox): /env 查看环境变量
   /linter 自动修复格式问题
   /test 执行测试，若有失败自动调整
'''
print(Claude_Code_Flow)
```

## 2.2 OpenCode 工作流

```python
OpenCode_Flow = '''
1. 启动: opencode 命令，默认使用 Claude 3.7 Sonnet
   --model gemini-2.0-flash 切换模型

2. 多模型支持: Claude/Gemini/Qwen/Ollama
   /model 动态切换 AI 平台

3. 代码生成: 基于指定目录的智能编译
   全局和局部代码约束

4. TUI 界面: 高密度字体显示
   鼠标实时点击代码行参对话
'''
print(OpenCode_Flow)
```

## 2.3 代码质量对比实验

```python
import anthropic, os

client = anthropic.Anthropic(api_key=os.environ.get('ANTHROPIC_API_KEY',''))

tasks = [
    ('简单函数',    'Write a Python function to check if a string is a palindrome.'),
    ('数据类型',    'Create a dataclass User with fields: id (int), name (str), email (str).'),
    ('错误处理',    'Write a function divide(a, b) that raises ValueError when b==0.'),
    ('测试生成',    'Write pytest tests for a stack with push/pop/peek/is_empty.'),
]

for title, prompt in tasks:
    r = client.messages.create(
        model='claude-sonnet-4-7-20252004',
        max_tokens=400,
        messages=[{'role': 'user', 'content': prompt}]
    )
    has_error = 'Traceback' in r.content[0].text[:100]
    print(f'[{"Y" if has_error else "N"}] {title}: {len(r.content[0].text)} chars')
```

## 2.4 实验结论
