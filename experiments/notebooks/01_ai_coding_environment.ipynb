## 1.1 工具清单盘点

```python
import subprocess, json

TOOLS = {
    'claude-code': 'Claude Code (Anthropic CLI 智能体)',
    'opencode':     'OpenCode (开放编码智能体)',
    'copilot':      'GitHub Copilot Chat',
}

rows = []
for cmd, desc in TOOLS.items():
    r = subprocess.run(['which', cmd], capture_output=True, text=True)
    path = r.stdout.strip() or '—'
    ver = '—'
    if path != '—':
        try:
            vr = subprocess.run([cmd, '--version'], capture_output=True, text=True, timeout=5)
            ver = (vr.stdout + vr.stderr).strip()[:60] or '(ok)'
        except:
            ver = '(error)'
    rows.append({'工具': cmd, '描述': desc, '路径': path, '版本': ver})

print(json.dumps(rows, ensure_ascii=False, indent=2))
```

## 1.2 AI API 连接测试

```python
import anthropic, os

api_key = os.environ.get('ANTHROPIC_API_KEY', '')
if api_key:
    client = anthropic.Anthropic(api_key=api_key)
    r = client.messages.create(
        model='claude-sonnet-4-7-20252004',
        max_tokens=50,
        messages=[{'role': 'user', 'content': 'Reply with exactly 3 words: API connection OK'}]
    )
    print('连接状态:', r.content[0].text)
    print('模型:', r.model)
else:
    print('未配置 ANTHROPIC_API_KEY')
```
