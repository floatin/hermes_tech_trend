## 4.2 分层 Review 策略\n\n### 三层审查\n\n```\n┌─────────────────────────────────────────────────────────┐\n│ Layer 1: 规范层（Spec Compliance）                      │\n│   AI 检查代码是否遵守设计规范和 API 契约                 │\n├─────────────────────────────────────────────────────────┤\n│ Layer 2: 安全层（Security & Safety）                   │\n│   AI 检查潜在的安全漏洞和危险操作                        │\n├─────────────────────────────────────────────────────────┤\n│ Layer 3: 质量层（Code Quality）                         │\n│   AI 检查代码风格、可读性、最佳实践                      │\n└─────────────────────────────────────────────────────────┘\n```\n\n### 关键词触发机制\n\n- **@设计规范**：AI 检查是否引用了 DESIGN.md 中的令牌\n- **@API契约**：AI 检查函数签名、参数类型、返回值是否与契约一致\n- **@安全**：AI 检查是否有注入、越权、数据泄露风险

## 4.4 AI Review 的局限性\n\n### AI 不擅长的领域\n\n- **架构设计**：AI 难以评估代码的架构是否合理\n- **业务逻辑**：AI 不理解业务背景，无法判断业务逻辑正确性\n- **性能优化**：AI 难以准确评估性能瓶颈\n- **团队规范**：AI 不了解特定团队的编码习惯\n\n### 人机协作的最佳实践\n\n1. **AI 做筛选**：AI 快速扫描，发现明显问题\n2. **人类做判断**：人类决定哪些问题需要修复\n3. **AI 做验证**：修复后 AI 再次检查

## 4.5 实验结论\n\n### AI Code Review 的核心价值\n- **速度**：快速扫描大量代码，发现模式化问题\n- **一致性**：相同规则下，结果稳定可复现\n- **全面性**：不受\"Review 疲劳\"影响\n- **可追溯**：每个问题都有对应的规则依据\n\n### 关键洞察\n- AI Review 的核心不是发现 bug，而是验证代码是否遵守规范\n- 规范是 AI Review 的前提：无规范，则无合规性检查\n- AI Review + Spec-Driven = 规范驱动的开发流程
